```python
import pandas as pd
import math
import numpy as np
```

```python
f_train = open("reviewstrain.txt", 'r')
lines = f_train.readlines()
token = []
token_count = []
for i in lines:
    c = 0
    for j in i.split():
        if c == 0:
            c += 1
            continue
        if j not in token:
            token.append(j)
            token_count.append(1)
        else:
            token_count[token.index(j)] += 1
        c += 1
```

```python
df = pd.DataFrame()
df['tokens'] = token
df['count'] = token_count
df.nlargest(5, 'count')
```

```python
tkn = token + ['label']
df1 = pd.DataFrame(columns=tkn)
df1
```

```python
t = [0] * len(tkn)
for i in lines:
    c = 0
    for j in i.split():
        if c == 0:
            t[len(tkn)-1] = int(j)
        else:
            t[token.index(j)] = 1
        c += 1
    tdf = pd.DataFrame([t], columns=tkn)
    df1 = df1.append(tdf, ignore_index=True, sort=False)
    t = [0] * len(tkn)
df1
```

```python
df1 = pd.read_excel('output.xlsx')
df1
```

```python
c0 = 0
c1 = 0
for item in df1['label']:
    if item == 1:
        c1 += 1
    else:
        c0 += 1
e = -(((c0/(c0+c1)) * math.log(c0/(c0+c1), 2)) + ((c1/(c0+c1)) * math.log(c1/(c0+c1), 2)))
```

```python
igs = []
for column in token:
    ct1 = 0
    ct0 = 0
    cf1 = 0
    cf0 = 0
    c0 = 0
    c1 = 0
    ct = 0
    cf = 0
    ig = 0
    for i,j in zip(df1[column],df1['label']):
        if i == 1:
            c1 += 1
            if j == 1:
                ct1 += 1
            else:
                ct0 += 1
        else:
            c0 += 1
            if j == 1:
                cf1 += 1
            else:
                cf0 += 1
    if (ct0/(ct0+ct1)) == 0:
        v1 = 0
    else:
        v1 = ((ct0/(ct0+ct1)) * math.log(ct0/(ct0+ct1), 2))
    if (ct1/(ct0+ct1)) == 0:
        v2 = 0
    else:
        v2 = ((ct1/(ct0+ct1)) * math.log(ct1/(ct0+ct1), 2))
    ct = -(v1 + v2)
    
    if (cf0/(cf0+cf1)) == 0:
        u1 = 0
    else:
        u1 = ((cf0/(cf0+cf1)) * math.log(cf0/(cf0+cf1), 2))
    if (cf1/(cf0+cf1)) == 0:
        u2 = 0
    else:
        u2 = ((cf1/(cf0+cf1)) * math.log(cf1/(cf0+cf1), 2))
    cf = -(u1 + u2)
    ig = e - (((c1/(c0 + c1)) * ct) + ((c0/(c0 + c1)) * cf))
    igs.append(ig)
```

```python
df2 = pd.DataFrame()
df2['token'] = token
df2['ig'] = igs
df3 = df2.nlargest(50, 'ig')
df3
```

```python
df4 = pd.DataFrame()
for i in df3['token']:
    df4[i] = df1[i]
df5 = pd.DataFrame()
df5['label'] = df1['label']
df4
```

```python
I = df4.values
D = df5.values
n_hidden = 4
eta = 0.01
n_max=1000
```

```python
def sigmoid(x):
    return (1 / (1 + np.exp(-(x.astype(float)))))
```

```python
np.random.seed(1926)
    
r_inp, c_inp = I.shape
n_examples = r_inp
n_input    = c_inp
r_out,c_out = D.shape
n_output   = c_out

w = np.random.random((n_input, n_hidden))
wb = np.random.random(n_hidden)
v = np.random.random((n_hidden, n_output))
vb = np.random.random(n_output)
err_curve = np.zeros((n_max,c_out))

for n in range(n_max):
#     sq_err_sum = np.zeros((1,n_output))
#     cross_e = 0

    for k in range(n_examples):
        x = I[k,:].reshape([1,-1])
        z = sigmoid(x.dot(w)+wb)
        y = sigmoid(z.dot(v)+vb)
        a = D[k,:]
        b = y
#         err = -(a*math.log(b) + (1-a)*math.log(1-b))
#         err = -(a*math.log(b, 2) + (1-a)*math.log(1-b, 2))
#         cross_e += err 
#         err = (D[k,:] - y).reshape([1,-1])
#         sq_err_sum += 0.5*np.square(err.astype(float))

#         Delta_output = err*(1-y)*y
        Delta_output = ((a/b) - ((1-a)/(1-b)))*(1-y)*y
#         Delta_output = (((a/b)*(1/math.log(2))) - (((1-a)/(1-b))*(1/math.log(2))))*(1-y)*y
        Delta_v = z.T.dot(Delta_output)
        Delta_vb = np.sum(Delta_output,axis=0)

        Delta_hidden = Delta_output.dot(v.T)*(1-z)*z
        Delta_w = x.T.dot(Delta_hidden)    
        Delta_wb = np.sum(Delta_hidden,axis=0) 
    
        v += eta*Delta_v.astype(float)
        vb += eta*Delta_vb.astype(float)
        w += eta*Delta_w.astype(float)
        wb += eta*Delta_wb.astype(float)

#         err_curve[n] = sq_err_sum/n_examples
#         err_curve[n] = cross_e/n_examples

#     print('epoch %d: err %f'%(n,np.mean(sq_err_sum)/n_examples))
#     print('epoch %d: err %f'%(n,cross_e/n_examples))
```
