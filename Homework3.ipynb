```python
import numpy.linalg as la
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import fetch_lfw_people
import sklearn.decomposition as skd
```

```python
#code for question 2 subpart d
a = np.matrix([[0, 14],[6, 9]])
w,v = la.eig(a)
w,v
```

```python
#Code for question 3 subpart c 
Y = np.matrix([[-3/4,-6/4,2/4],[13/4,10/4,2/4],[5/4,-10/4,-14/4],[-16/4,6/4,10/4]])
np.cov(Y.transpose())
```

```python
#code for question 3 subpart d
pca = skd.PCA(n_components = 3)
skd.PCA.fit(pca,Y)
W1 = pca.components_
W = W1.transpose()
Z = pca.transform(Y)
print("W : ", W)
print("Z : ", Z)
```

```python
#code for question 5
lfw_people = fetch_lfw_people(min_faces_per_person=70)
n_samples, h, w = lfw_people.images.shape
fea = lfw_people.data
plt.figure(figsize=(10,20))
nplt = 4
for i in range(nplt):
    plt.subplot(1,nplt,i+1)
    plt.imshow(fea[i].reshape((h, w)), cmap=plt.cm.gray)
```

```python
#code for question 5 subpart a
plt.imshow(fea[3].reshape((h, w)), cmap=plt.cm.gray)
```

```python
#code for question 5 subpart b
M = np.mean(fea, axis = 0)
plt.imshow(M.reshape((h, w)), cmap=plt.cm.gray)
```

```python
#code for question 5 subpart c
pca = skd.PCA(n_components = 5)
skd.PCA.fit(pca,fea)
W1 = pca.components_
W = W1.transpose()
Z = pca.transform(fea)
print(Z[3])
```

```python
#code for question 5 subpart d with 5 pca components
WT = np.matrix(W).transpose()
Z = np.matrix(Z)
X = np.matmul(Z, WT)
X = X + M
plt.imshow(X[3].reshape((h, w)), cmap=plt.cm.gray)
```

```python
#code for question 5 subpart d with 50 pca components
pca = skd.PCA(n_components = 50)
skd.PCA.fit(pca,fea)
W1 = pca.components_
W = W1.transpose()
Z = pca.transform(fea)
WT = np.matrix(W).transpose()
Z = np.matrix(Z)
X = np.matmul(Z, WT)
X = X + M
plt.imshow(X[3].reshape((h, w)), cmap=plt.cm.gray)
```
