```python
import pandas as pd
import math
import statistics as s
```

```python
spam_ham_df = pd.read_csv("spambasetrain.csv", header=None)
spam_ham_df.head()
```

```python
ham_df = spam_ham_df[spam_ham_df[9] == 0]
ham_df.index = pd.RangeIndex(len(ham_df.index))
ham_df.describe()
```

```python
spam_df = spam_ham_df[spam_ham_df[9] == 1]
spam_df.index = pd.RangeIndex(len(spam_df.index))
spam_df.describe()
```

```python
ham = len(ham_df)/len(spam_ham_df)
spam = len(spam_df)/len(spam_ham_df)
print(ham)
print(spam)
```

```python
means_ham = []
means_spam = []
for i in range(0, 10):
    means_ham.append(ham_df[i].mean())
    means_spam.append(spam_df[i].mean())
print(means_ham)
print(means_spam)
```

```python
variances_ham = []
variances_spam = []
sh = 0
ss = 0
for i in range(0, 10):
    for j in range(0, len(ham_df)):
        sh = sh + ((ham_df[i][j] - means_ham[i]) ** 2)
    for k in range(0, len(spam_df)):
        ss = ss + ((spam_df[i][k] - means_spam[i]) ** 2)
    variances_ham.append(sh / (len(ham_df) - 1))
    variances_spam.append(ss / (len(spam_df) - 1))
    sh = 0
    ss = 0
print(variances_ham)
print(variances_spam)
```

```python
v_ham = []
v_spam = []
for i in range(0, 10):
    a = list(ham_df[i])
    b = list(spam_df[i])
    v_ham.append(s.stdev(a))
    v_spam.append(s.stdev(b))
print(v_ham)
print(v_spam)
```

```python
spam_ham_df_test = pd.read_csv("spambasetest.csv", header=None)
spam_ham_df_test.head()
```
